# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### --- 1. Basic Line Plot ---

```python

print("\n1. Basic Line Plot")
print("--------------------")
# Generate some data
x = np.linspace(0, 10, 100) # 100 points between 0 and 10
y = np.sin(x)

# Create a figure and an axes (plt.subplots() is recommended for this)
fig, ax = plt.subplots(figsize=(8, 4)) # fig is the whole window, ax is the plotting area

# Plot the data
ax.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Add title and labels
ax.set_title('Simple Sine Wave Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a legend
ax.legend()

# Add a grid
ax.grid(True)

# Show the plot
plt.tight_layout() # Adjusts plot to prevent labels from overlapping
plt.show()
```

### --- 2. Scatter Plot ---

```python

print("\n2. Scatter Plot")
print("---------------")
# Generate some random data
np.random.seed(42) # for reproducibility
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
colors = np.random.rand(50)
sizes = np.random.rand(50) * 200 # marker sizes

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.7, cmap='viridis', label='Random Points')

ax.set_title('Scatter Plot with Color and Size Variation')
ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

### --- 3. Bar Plot ---

```python

print("\n3. Bar Plot")
print("-----------")
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 45, 56, 12]

fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

ax.set_title('Bar Chart of Categories')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
# Rotate x-axis labels if they are long
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### --- 4. Histogram ---

```python

print("\n4. Histogram")
print("------------")
# Generate random data from a normal distribution
data_hist = np.random.randn(1000) * 2 + 5 # Mean 5, Std Dev 2

fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(data_hist, bins=30, color='purple', alpha=0.7, edgecolor='black')

ax.set_title('Histogram of Random Data')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### --- 5. Subplots (Multiple Plots in One Figure) ---

```python

print("\n5. Subplots")
print("-----------")
# Using plt.subplots() for a grid of plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) # 2 rows, 2 columns

# Plot 1: Top-left
axes[0, 0].plot(x, y**2, color='red')
axes[0, 0].set_title('Y = sin(X)^2')
axes[0, 0].grid(True)

# Plot 2: Top-right
axes[0, 1].scatter(x_scatter, y_scatter, color='green', alpha=0.5)
axes[0, 1].set_title('Random Scatter')

# Plot 3: Bottom-left
axes[1, 0].hist(data_hist, bins=20, color='orange', alpha=0.8)
axes[1, 0].set_title('Histogram of Data')

# Plot 4: Bottom-right
axes[1, 1].barh(categories, values, color='teal') # Horizontal bar chart
axes[1, 1].set_title('Horizontal Bar Chart')

plt.suptitle('Multiple Plots using Subplots', fontsize=16) # Title for the entire figure
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout, leaving space for suptitle
plt.show()
```


### --- 6. Customization: Colors, Markers, Linestyles, Text, Annotations ---

```python

print("\n6. Customization and Annotations")
print("----------------------------------")
fig, ax = plt.subplots(figsize=(9, 6))

x_custom = np.linspace(0, 5, 50)
y_custom1 = x_custom**2
y_custom2 = np.exp(x_custom / 2)

ax.plot(x_custom, y_custom1, label='Quadratic ($x^2$)', color='darkblue',
        linestyle='--', marker='o', markersize=5, markerfacecolor='lightblue',
        markeredgecolor='darkblue', markeredgewidth=1)

ax.plot(x_custom, y_custom2, label='Exponential ($e^{x/2}$)', color='darkgreen',
        linestyle=':', marker='x', markersize=6, linewidth=2)

ax.set_title('Customized Plot with Annotations', fontsize=16, color='darkred')
ax.set_xlabel('Independent Variable', fontsize=12, color='gray')
ax.set_ylabel('Dependent Variable', fontsize=12, color='gray')
ax.legend(fontsize=10, loc='upper left', frameon=True, shadow=True, fancybox=True)

# Set limits
ax.set_xlim(0, 5.5)
ax.set_ylim(0, 15)

# Add text directly on the plot
ax.text(3, 10, 'Important Point', fontsize=10, color='brown', weight='bold')

# Add an annotation (arrow + text)
ax.annotate('Local Min', xy=(0, 0), xytext=(1, 2),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=9, color='black')

# Change tick parameters
ax.tick_params(axis='both', which='major', labelsize=10, length=6, width=1.5, colors='gray')

plt.tight_layout()
plt.show()
```
