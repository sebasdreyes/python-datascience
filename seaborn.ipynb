# Seaborn

### Import libraries and set a style

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set a style for all plots (optional, but good practice for aesthetics)
sns.set_theme(style="whitegrid")
```

### --- 1. Loading Built-in Datasets ---

```python
print("\n1. Loading Built-in Datasets")
print("----------------------------")
# Seaborn comes with several useful datasets for examples
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
flights = sns.load_dataset("flights")

print("First 5 rows of the 'tips' dataset:")
print(tips.head())
print("\nFirst 5 rows of the 'iris' dataset:")
print(iris.head())
```

### --- 2. Relationship Plots ---

```python
print("\n2. Relationship Plots")
print("---------------------")

# Scatter plot using relplot (flexible and robust)
# By default, relplot uses scatter. Can map categories to hue, size, style.
print("\nScatter Plot (relplot with hue and size):")
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", size="size", sizes=(20, 200),
            col="time", col_wrap=2, height=4, aspect=1.2)
plt.suptitle('Total Bill vs Tip by Smoker and Time', y=1.02) # Adjust suptitle position
plt.show()

# Line plot using lineplot (for showing trends, especially with aggregation)
print("\nLine Plot (lineplot for trends):")
# Aggregates by default (mean and confidence interval)
sns.lineplot(data=flights, x="year", y="passengers", hue="month", style="month",
             palette="viridis", errorbar=("se", 2)) # show standard error band
plt.title('Passengers over Years, by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### --- 3. Distribution Plots ---

```python

print("\n3. Distribution Plots")
print("---------------------")

# Histogram with KDE (Kernel Density Estimate)
print("\nHistogram (histplot with KDE):")
sns.histplot(data=tips, x="total_bill", kde=True, bins=20, color="skyblue")
plt.title('Distribution of Total Bill')
plt.tight_layout()
plt.show()

# KDE Plot
print("\nKDE Plot (kdeplot for density estimation):")
sns.kdeplot(data=tips, x="tip", fill=True, color="green")
plt.title('Density Estimation of Tip Amount')
plt.tight_layout()
plt.show()

# Displot (flexible distribution plot - can be hist, kde, ecdf)
print("\nDisplot (flexible distribution plot, here as histogram):")
sns.displot(data=tips, x="total_bill", col="time", row="smoker", bins=15, kde=True,
            height=3, aspect=1.5)
plt.suptitle('Distribution of Total Bill by Time and Smoker', y=1.02)
plt.tight_layout()
plt.show()
```

### --- 4. Categorical Plots ---

```python
print("\n4. Categorical Plots")
print("--------------------")

# Box Plot
print("\nBox Plot (boxplot to show distribution across categories):")
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker", palette="coolwarm")
plt.title('Total Bill by Day and Smoker (Box Plot)')
plt.tight_layout()
plt.show()

# Violin Plot (combines box plot with KDE)
print("\nViolin Plot (violinplot showing density and quartiles):")
sns.violinplot(data=tips, x="day", y="tip", hue="time", split=True, inner="quartile",
               palette="pastel")
plt.title('Tip Amount by Day and Time (Violin Plot)')
plt.tight_layout()
plt.show()

# Bar Plot (shows mean of a numerical variable for each category)
print("\nBar Plot (barplot for comparing mean values):")
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", estimator=np.mean, ci="sd",
            palette="viridis") # ci="sd" for standard deviation error bars
plt.title('Average Total Bill by Day and Sex')
plt.tight_layout()
plt.show()

# Count Plot (shows counts of observations in each category)
print("\nCount Plot (countplot for categorical variable counts):")
sns.countplot(data=tips, x="day", hue="smoker", palette="magma")
plt.title('Number of Customers by Day and Smoker')
plt.tight_layout()
plt.show()
```

### --- 5. Regression Plots ---

```python
print("\n5. Regression Plots")
print("-------------------")

# Regplot (shows scatter plot with linear regression model fit)
print("\nRegplot (scatter plot with linear regression line):")
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
plt.title('Total Bill vs Tip with Regression Line')
plt.tight_layout()
plt.show()

# Lmplot (combines regplot with FacetGrid for plotting across subsets)
print("\nLmplot (regression plot across subsets):")
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", row="smoker", hue="sex",
           height=4, aspect=1.2, scatter_kws={"alpha":0.6})
plt.suptitle('Total Bill vs Tip Regression by Time, Smoker, and Sex', y=1.02)
plt.tight_layout()
plt.show()
```

### --- 6. Matrix Plots (Heatmaps) ---

```python

print("\n6. Matrix Plots (Heatmaps)")
print("--------------------------")

# Heatmap for correlation matrix
print("\nHeatmap of Correlation Matrix:")
# Select only numeric columns for correlation calculation
numeric_tips = tips.select_dtypes(include=np.number)
correlation_matrix = numeric_tips.corr(numeric_only=True) # Ensure only numeric columns are used for correlation

plt.figure(figsize=(7, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Tips Dataset')
plt.tight_layout()
plt.show()

# Heatmap for pivot table data (e.g., flight passengers by year and month)
print("\nHeatmap of Flight Passengers (Pivot Table):")
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
plt.figure(figsize=(9, 7))
sns.heatmap(flights_pivot, cmap="Blues", annot=True, fmt="d", linewidths=.5, cbar_kws={'label': 'Passengers'})
plt.title('Flight Passengers by Month and Year')
plt.tight_layout()
plt.show()
```


### --- 7. Multi-plot Grids (Advanced) ---

```python
print("\n7. Multi-plot Grids (Advanced)")
print("------------------------------")

# Pairplot (scatter plots for all pairwise relationships)
print("\nPairplot (pairwise relationships):")
# Exclude 'day' column if it's not numeric, pairplot automatically handles hue
sns.pairplot(iris, hue="species", palette="dark")
plt.suptitle('Pairwise Relationships in Iris Dataset', y=1.02)
plt.show()

# FacetGrid (more custom control over subplots)
print("\nFacetGrid (custom multi-plot grid):")
g = sns.FacetGrid(tips, col="time", row="smoker", hue="sex", height=3, aspect=1.5, palette="tab10")
g.map(sns.scatterplot, "total_bill", "tip", alpha=0.7) # Map a plotting function to each facet
g.add_legend()
g.set_axis_labels("Total Bill ($)", "Tip ($)")
g.set_titles(col_template="{col_name} Time", row_template="{row_name} Smoker")
plt.suptitle('Total Bill vs Tip by Time, Smoker, and Sex (FacetGrid)', y=1.02)
plt.tight_layout()
plt.show()


print("\n--- End of Seaborn Script ---")
```
